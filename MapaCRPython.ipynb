```python
from folium.plugins import MiniMap
from branca.element import Figure

popuptext = '<b>Guanacaste</b>'

my_map = folium.Map(location=[10.2856822,-84.9008912], zoom_start=16,titles='stamenterrain')
folium.Marker(location=[10.2856822,-84.9008912],popup=popuptext, icon=folium.Icon(color="orange",icon="ok-sign")).add_to(my_map)
folium.Circle(location=[10.2856822,-84.9008912],color="purple",weight=6,fill_color="red",fill_opacity=0.5,tooltip="Gunacaste").add_to(my_map)
minimap = MiniMap(title_layer='stamenterrain', position='bottomright')
my_map.add_child(minimap)
```

```python
# Marcando 5 sedes de la Universidad Nacional de Costa Rica a lo largo del país

# Coordenadas de las ubicaciones
locations = [
    {"name": "Campus Omar Dengo", "lat": 9.998868412136458, "lon": -84.11147425226446},
    {"name": "Campus Liberia", "lat": 10.617665011856845, "lon": -85.45147164960504},
    {"name": "Campus Pérez Zeledón", "lat": 9.37988387151278, "lon": -83.69152493427977},
    {"name": "Campus Benjamín Núñez", "lat": 9.969903003813792, "lon": -84.12902437411505},
    {"name": "Campus Sarapiquí", "lat": 10.319318883813203, "lon": -83.92283293360971}
]

# Crear el mapa
my_map = folium.Map(location=[9.998868412136458, -84.11147425226446], zoom_start=8, tiles='stamenterrain')

# Agregar marcadores y círculos para cada ubicación
for location in locations:
    popuptext = f'<b>{location["name"]}</b>'
    folium.Marker(location=[location["lat"], location["lon"]], popup=popuptext, icon=folium.Icon(color='orange', icon='ok-sign')).add_to(my_map)
    folium.Circle(location=[location["lat"], location["lon"]], color='purple', weight=6, fill_color='red', fill_opacity=0.5, tooltip=location["name"]).add_to(my_map)

# Crear el minimapa
minimap = MiniMap(title_layer='stamenterrain', position='bottomright')
my_map.add_child(minimap)

# Guardar el mapa en un archivo HTML o mostrarlo
my_map.save("sedes_una.html")
my_map
```
